# 并行实验与评估演示
本Notebook演示如何利用 `run_experiment.py` 脚本在后台并行运行多个实验，并自动收集结果。
这对于对比不同算法（如 QCPO vs QPO）或不同超参数非常有用。

```python
import subprocess
import time
import sys
import os
import glob
import json
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# 定义实验配置
experiments = [
    {"algo_name": "QCPO", "seed": 0, "device": "0"},
    {"algo_name": "QCPO", "seed": 1, "device": "0"},
    {"algo_name": "QPO", "seed": 0, "device": "0"},
    {"algo_name": "QPO", "seed": 1, "device": "0"},
]

# 并行运行的最大进程数
MAX_PARALLEL = 16  # 根据你的显存和CPU核心数调整
OUTPUT_DIR = "results"

processes = []

for exp in experiments:
    cmd = [
        sys.executable, "run_experiment.py",
        "--algo_name", exp["algo_name"],
        "--seed", str(exp["seed"]),
        "--device", exp["device"],
        "--output_dir", OUTPUT_DIR,
        "--max_episode", "1000",  # 演示用，减少训练步数
        "--eval_episodes", "100"  # 演示用，减少评估步数
    ]
    
    # 简单的进程调度
    while len(processes) >= MAX_PARALLEL:
        # 移除已结束的进程
        processes = [p for p in processes if p.poll() is None]
        if len(processes) >= MAX_PARALLEL:
            time.sleep(1)
            
    print(f"Starting experiment: {exp}")
    p = subprocess.Popen(cmd)
    processes.append(p)

# 等待所有进程结束
for p in processes:
    p.wait()
    
print("All experiments finished!")
```

```python
# 读取并分析结果
result_files = glob.glob(os.path.join(OUTPUT_DIR, "*.json"))
data = []

for file in result_files:
    with open(file, 'r') as f:
        res = json.load(f)
        data.append({
            "Algorithm": res["algo_name"],
            "Seed": res["seed"],
            "Mean Return": res["mean_est"],
            "Quantile Return": res["quantile_est"]
        })

df = pd.DataFrame(data)
print(df)

# 简单的条形图对比
if not df.empty:
    df.groupby("Algorithm")[["Mean Return", "Quantile Return"]].mean().plot(kind="bar")
    plt.title("Algorithm Comparison")
    plt.ylabel("Return")
    plt.show()
```
